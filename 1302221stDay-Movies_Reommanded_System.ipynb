```python

import ast
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.iloc[0].genres
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
movies.iloc[0].keywords
```

```python
# frist 3 Actors name Count
def convert3(obj):
  L=[]
  count=0
  for i in ast.literal_eval(obj):
    if count<3:
      L.append(i['name'])
      count+=1
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.iloc[0].cast
```

```python
movies.head()
```

```python
# Only Director Name Assain

def convertDic(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
  return L
```

```python
movies['crew']=movies['crew'].apply(convertDic)
```

```python
movies.iloc[0].crew
```

```python
movies.head()
```

```python
# Every Multi word name collapse of flowindly Indentify.
def collapse(obj):
  L=[]
  for i in obj:
    L.append(i.replace(" ",""))
  return L
```

```python

movies['genres']=movies['genres'].apply(collapse)
```

```python
movies['keywords']=movies['keywords'].apply(collapse)
movies['cast']=movies['cast'].apply(collapse)
movies['crew']=movies['crew'].apply(collapse)
```

```python
movies.head()
```

```python
x=movies.iloc[2].overview
```

```python
# avery word user of list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_movies=movies.drop(['overview','genres','keywords','cast','crew'],axis='columns')
```

```python
new_movies.head()
```

```python
x=new_movies.iloc[0].tags
print(x)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new_movies['tags']).toarray()
```

```python
vector.shape
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python


similarity = cosine_similarity(vector)

```

```python
similarity
```

```python
new_movies.iloc[0].title
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=(lambda x: x[1]))[1:6]
```

```python
def recommend(movie):
  movies_index=new_movies[new_movies['title']==movie].index[0]
  distance=similarity[movies_index]
  movies_list=sorted(list(enumerate(distance)),reverse=True,key=(lambda x: x[1]))[1:6]
  for i in movies_list:
    print(new_movies.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
